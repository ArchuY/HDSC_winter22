```python
# Import required Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Read the energy data dataset

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
energy_data = pd.read_csv(url, encoding = 'latin-1')
```

### Question 12

#### From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
# split the data for train and test the model
x_train,x_test,y_train,y_test = train_test_split(energy_data[['T2']],energy_data[['T6']],test_size = 0.3, random_state= 42)

# Linear regression model
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_value = linear_model.predict(x_test)

# Calculating the R-squared score
r_sqr  = r2_score(y_test,predicted_value)
print('The R-squared value upto two decimal point is:',round(r_sqr,2))
```

###Question 13

#### Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. 
#### The target variable is “Appliances”. 
#### Use a 70-30 train-test set split with a random state of 42 (for reproducibility). 
#### Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:
#### What is the Mean Absolute Error (in two decimal places)?


```python
# Drop the Date and lights columns from dataset
energy_data.drop(['date','lights'], axis = 1, inplace = True)
energy_data.head()
```

```python
# Normalization using MinMaxScaler
scaler = MinMaxScaler()
normalized = pd.DataFrame(scaler.fit_transform(energy_data),columns = energy_data.columns)
normalized.head()
```

```python
# Making Appliances column as target variable
feature = normalized.drop('Appliances', axis = 1)
target = normalized['Appliances']
print("Target variable:\n",target.head(4))


# Splitting dataset into 70-30 training and testing dataset with a random state of 42 
x_train,x_test,y_train,y_test = train_test_split(feature,target,test_size = 0.3, random_state = 42)
print('\n\nTrain sample:',x_train.shape,'| Test sample:',x_test.shape)
```

```python
#multiple linear regression using the training set and evaluate your model on the test set.
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_values = linear_model.predict(x_test)

#Evaluate model by Mean Absolute Error 
mae = mean_absolute_error(y_test,predicted_values)
print('The Mean Square Error upto two decimal point is:',round(mae,2))
```

### Question 14
#### What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.square(y_test - predicted_values))
print('Residual Sum of Square upto two decimal point is:', round(rss,2))
```

### Question 15
#### What is the Root Mean Squared Error (in three decimal places)?

```python
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
print('The Root Mean Square Error upto three decimal places is:',round(rmse,3))
```

### Question 16
#### What is the Coefficient of Determination (in two decimal places)?

```python
cod = r2_score(y_test,predicted_values)
print('Coefficent of Determination (R-squared) in two decimal point is:', round(cod,2))
```

### Question 17
#### Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
# finding the weights for all features from linear model
coeff = pd.DataFrame(linear_model.coef_,feature.columns, columns =['Coefficient'])

# finding feature with lowest weight
print("features have the lowest weights is:\n ",coeff['Coefficient'].nsmallest(1))

# finding feature wwith highest weight
print("\n\nfeatures have the highest weights is:\n ",coeff['Coefficient'].nlargest(1))
```

### Question 18
#### Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?


```python
# Ridge Regression
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train,y_train)
predicted_ridge = ridge_reg.predict(x_test)
rmse_ridge = np.sqrt(mean_squared_error(y_test,predicted_ridge))
print('The new rmse:',round(rmse_ridge,3))
```

No, there is no changes to the root mean squared error (RMSE) when evaluated on the test *set*

### Question 19
#### Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
#Lasso Regression
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train,y_train)
predicted_lasso = lasso_reg.predict(x_test)
#New feature weights from Lasso regression model
lasso_coeff = pd.DataFrame(lasso_reg.coef_,feature.columns, columns =['Lasso_Coefficient'])

# Finding non zero feature weights
print('The number of features have non-zero feature weights are: ',(lasso_coeff['Lasso_Coefficient']!= 0).sum())
```

### Question 20

#### What is the new RMSE with the lasso regression? (Answer should be in three (3) decimal places)



```python
# Root Mean Squared Error with Lasso regression model
rmse_lasso = np.sqrt(mean_squared_error(y_test, predicted_lasso))
print('New RMSE with the lasso regression is :',round(rmse_lasso,3))
```
